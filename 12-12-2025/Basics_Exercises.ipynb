6. Create a new column named "spend_lakh" = annual_spend / 100000.

```python
from pyspark.sql.functions import col
df2=df.withColumn("spend_lakh", col('annual_spend')/100000)
df2.show()
```

7. Create a new column "customer_type"
Logic:
spend > 70000 → Premium
else → Standard
Use when() and otherwise().

```python
from pyspark.sql.functions import col,when
df3=df.withColumn('customer_type',when (col("annual_spend")>70000, "Premium").otherwise ("Standard")).show()
```

13. Find customers who are younger than the average age of the entire dataset.
Steps:
Calculate average age.
Filter df where age < avg_age.

```python
from pyspark.sql.functions import avg, first
avg_age = df.select(avg("age").alias("avg_age")).first()[0]
df.filter(col("age") < avg_age)
```
